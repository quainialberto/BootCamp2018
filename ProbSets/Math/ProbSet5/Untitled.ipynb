### Import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

Define function to compute intersection

```python
def intersect(a1, b1, c1, a2, b2, c2):
    """a1x + b1y = c1
       a2x + b2y = c2
    """
    determinant = a1 * b2 - a2 * b1
    if (determinant == 0):
        print("Lines are parallel")

    else:
        x = (c1 * b2 - c2 * b1)/determinant
        y = (a1 * c2 - a2 * c1)/determinant
    return x, y
```

## Exercise 1

```python
x = np.linspace(0,4,100)

# The lines to plot
y1 = 4/3 + 2/3 * x
y2 = -1/6 + 1/6 * x
y3 = 6 - x
y = np.linspace(0,6,100)
X, Y = np.meshgrid(x, y)
z = 5*X - 4*Y

plt.ylim(0, 6)

plt.plot(x, y1,
         x, y2,
         x, y3)
plt.contour(x, y, z)
px, py = intersect(2, -3, -4, 1, 1, 6)
plt.plot(px, py, 'o')

plt.fill_between(x, y1, y3, color='grey', where = y3 >= y1, alpha='0.5')
plt.show()
print("Optimizer is: (", px, ", ", py, ")")
```
