## Problem 2

```python
# define the Bernstein function
def berny(x, n, v):
    return special.binom(n, v) * x**v * (1 - x)**(n - v)
```

```python
# 
plt.figure(figsize=(8, 8))
plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('tab20b')

x_grid = np.linspace(0, 1, 100)
for n in range(4):
    for v in range(n + 1):
        place = (n + 1) * 4 - 3 + v
        y = berny(x_grid, n, v)
        plt.subplot(4, 4, place)
        plt.plot(x_grid, y, color = palette(place), alpha=0.9)
        plt.title((n, v))
        plt.axis([0, 1, 0, 1])
        plt.tight_layout()

        
```

## Problem 4

```python
year, magnitude, longitude, latitude = np.load("earthquakes.npy").T
n = len(year)

plt.plot(year, magnitude, '.')
plt.xlabel("Year")
plt.ylabel("Magnitude")
plt.show()

year_floor = np.floor(year)
year_unique = np.unique(year_floor)
year_freq = np.zeros(len(year_unique))
for i in range(len(year_unique)):
    year_freq[i] = np.sum(year_floor == year_unique[i])
year_freq

plt.bar(year_unique, year_freq)
plt.xlabel('year')
plt.ylabel('Number earthquakes')
plt.title('Earthquakes per year')
plt.show()
```

```python
plt.hist(magnitude, bins = 30)
plt.xlabel('magnitude')
plt.ylabel('frequency')
plt.title('Magnitude frequencies')
plt.show()
```

```python
plt.scatter(latitude, longitude, 10, c=magnitude, cmap = 'magma')
plt.xlabel('latitude')
plt.ylabel('longitude')
cbar = plt.colorbar()
cbar.set_label("magnitude", labelpad = + 1)
plt.axis('equal')
plt.show()
```

## Problem 5

```python
@np.vectorize
def rosen(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2
```

```python
n = 100
grid = np.linspace(0, 2, n)
X, Y = np.meshgrid(grid, grid)
F = rosen(X, Y)

plt.pcolormesh(X, Y, F, cmap="plasma")
plt.plot(1, 1, 'g*', markersize = 10)
plt.colorbar()
plt.show()            
plt.contour(X, Y, F, 10, cmap="magma")
plt.plot(1, 1, 'g*', markersize = 10)
plt.colorbar()
plt.axis([0, 2, 0, 2])
plt.show()            
```

```python
data = np.load('countries.npy')
population = data[:,0]
gdp = data[:,1]
height_m = data[:,2]
height_f = data[:,3]
countries = ["Austria", "Bolivia", "Brazil", "China",
            "Finland", "Germany", "Hungary", "India",
            "Japan", "North Korea", "Montenegro", "Norway",
            "Peru", "South Korea", "Sri Lanka", "Switzerland",
            "Turkey", "United Kingdom", "United States", "Vietnam"]
```

```python
fig, ax = plt.subplots()
for i in range(len(countries)):
    ax.scatter(population[i], gdp[i], label = countries[i])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel('Population')
plt.ylabel('GDP')
plt.show()

```

```python
fig, ax = plt.subplots()
for i in range(len(countries)):
    ax.scatter(height_m[i], height_f[i], label = countries[i])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel('Height male')
plt.ylabel('Height female')
plt.show()
```
