### Import libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
from pydataset import data
import numpy as np
```

```python
LB = data('Arbuthnot')
LB.set_index('Year', inplace=True)
LB
```

```python
LB[['Males', 'Females']].plot()
plt.title('Births per year in London')
plt.grid(True)
plt.show()
```

There were more males born in each year than females over the time period

```python
LB['Plague'].plot()
plt.title('Deaths due to plague')
plt.grid(True)
plt.show()
```

In 1665 happaned the Great Plague of London

```python
log_mortality = np.log(LB['Mortality'] + 1)
log_plague = np.log(LB['Plague'] + 1)
plt.scatter(log_mortality, log_plague)
plt.title('Mortality vs Plague | log - log')
plt.xlabel('log Plague')
plt.ylabel('log Mortality')
plt.grid(True)
plt.show()
```

Plague seem to be an explanatory variable for Mortality.
