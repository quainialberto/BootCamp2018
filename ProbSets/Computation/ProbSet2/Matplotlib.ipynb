### Import libraries

```python
# Import
import numpy as np
from matplotlib import pyplot as plt
```

### Define functions to generate and plot variances

```python
# sample size
@np.vectorize
def variances(n):
    # generate nxn sample from gaussian distribution
    sample = np.random.normal(size = (n, n))
    # compute row-wise mean
    row_means = np.mean(sample, axis = 1)
    # compute the variance of these means
    return np.var(row_means)

def plot_variances():
    # sample size grid
    grid = np.arange(100, 1100, 100)
    var = variances(grid)
    # plot variances
    plt.plot(grid, var)
    plt.grid(True)
    plt.title('Variance of row-means')
    plt.xlabel('Sample size')
    plt.show()
```

### Plot variances

```python
plot_variances()
```

## Problem 2

```python
# Define function

def plot_trig():
    pi   = np.pi
    grid = np.linspace(-2*pi, 2*pi, 100)
    sins = np.sin(grid)
    coss = np.cos(grid)
    arctans = np.arctan(grid)
    plt.plot(grid, sins, label = "sin")
    plt.plot(grid, coss, label = "cos")
    plt.plot(grid, arctans, label = "arctan")
    plt.grid(True)
    legend = plt.legend(loc='lower right', shadow=False)
    for label in legend.get_texts():
        label.set_fontsize('large')
    plt.title('Trigonometric functions')
    plt.show()
```

### Plot trigonometric functions

```python
plot_trig()
```

## Problem 3

```python
# Define function

def plot_f():
    l_grid = np.linspace(-2, 1, 101)
    r_grid = np.linspace(1, 6, 101)
    l_f = 1 / (l_grid[:-1] - 1)
    r_f = 1 / (r_grid[1:] - 1)
    l_line, = plt.plot(l_grid[:-1], l_f, color = 'm', linewidth = 4)
    r_line, = plt.plot(r_grid[1:], r_f, color = 'm', linewidth = 4)
    plt.xlim(-2, 6)
    plt.ylim(-6, 6)
    l_line.set_dashes([6, 3])
    r_line.set_dashes([6, 3])
    plt.show()
```

### Plot function

```python
plot_f()
```

### Problem 4

```python
def problem4():
    pi = np.pi
    grid = np.linspace(0, 2 * pi)
    lims = [0, 2 * pi, -2, 2]
    
    fig = plt.figure()
    
    ax1 = fig.add_subplot(221)
    ax1.plot(grid, np.sin(grid), 'g')
    ax1.axis(lims)
    ax1.set_title('sin(x)')
    
    ax2 = fig.add_subplot(222)
    ax2.plot(grid, np.sin(2 * grid), 'r--')
    ax2.axis(lims)
    ax2.set_title('sin(2x)')

    ax3 = fig.add_subplot(223)
    ax3.plot(grid, 2 * np.sin(grid), 'b--')
    ax3.axis(lims)
    ax3.set_title('2sin(x)')

    ax4 = fig.add_subplot(224)
    ax4.plot(grid, 2 * np.sin(2 * grid), 'm:')
    ax4.axis(lims)
    ax4.set_title('2sin(2x)')
    
    plt.suptitle('Trigonometric functions')
    
    plt.tight_layout()
    #fig = plt.gcf()
    
    plt.show()
```

### Plot trigonometric functions

```python
problem4()
```

## Problem 5

### Define Function to visualize FARS

```python
def fars():
    data = np.load('FARS.npy')
    
    fig = plt.figure()
    
    ax1 = fig.add_subplot(211)
    ax1.plot(data[:, 1], data[:, 2], 'k,')
    ax1.set_xlabel('latitude')
    ax1.set_ylabel('longitude')
    #plt.axis("equal")
    
    ax2 = fig.add_subplot(212)
    ax2.hist(data[:, 0], 24)
    ax2.set_xlim([0, 23])
    ax2.set_xlabel('hour')
    ax2.set_ylabel('frequency')
    
    plt.tight_layout()
    
    plt.show()
```

```python
fars()
```

## Problem 6

```python
# Define function

def problem6():
    pi = np.pi
    n = 100
    grid = np.linspace(-2*pi, 2*pi, n)
    X, Y = np.meshgrid(grid, grid)
    G = np.sin(X) * np.sin(Y) / (X * Y)
    
    #fig = plt.figure()
    plt.subplot(211)
    plt.pcolormesh(X, Y, G, cmap="plasma")
    plt.colorbar()
    
    plt.subplot(212)
    plt.contour(X, Y, G, 10, cmap="magma")
    plt.colorbar()
    
    plt.axis([-2*pi, 2*pi, -2*pi, 2*pi])
    plt.tight_layout()
    plt.show()            
```

### Plot 3D plots

```python
problem6()
```
